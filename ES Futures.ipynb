### Import Libraries

```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
%matplotlib

ES = yf.download("ES=F", start="2024-01-01", end="2024-12-06")
ES["log_return"] = np.log(ES["Adj Close"] / ES["Adj Close"].shift(1))
```

```python
#Pull Price Data From Yahoo Finance

ES["day_of_month"] = ES.index.day
ES["year"] = ES.index.year
grouped_by_day = ES.groupby("day_of_month").log_return.mean()
grouped_by_day.plot.bar(title=" ES Mean Log Returns by Calendar Day of Month")
plt.show()
```
